```python
import pandas as pd 
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing as p
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
```

```python
columns_name=['Pregnancies','GlucosePlasma','Blood Pressure','SkinThickness','Insulin','BMI','Diabetes Pedigree Function','Age','Outcome']
```

```python
df.columns=columns_name
```

```python
df
```

```python
df.info()
```

```python
X = df.drop('Outcome',axis=1)
Y = df[['Outcome']]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=1,  test_size=0.20, shuffle=True)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
min_max_scaler = p.MinMaxScaler()
X_train_normalizedData = min_max_scaler.fit_transform(X_train)

X_test_normalized = min_max_scaler.transform(X_test)
```

```python
X_train_normalizedData
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_train_normalizedData, Y_train)
print(LR.score(X_test, Y_test))
```

```python
X_test.shape,Y_test.shape
```

```python

y_pred = LR.predict(X_test_normalized)

# Data scatter of predicted values
```

```python
y_pred
```

```python
y_hat = []
def round_output(y):
    for eachPrediction in y:
        if eachPrediction > 0.4:
            y_ = 1
        else:
            y_= 0
        
        y_hat.append(y_)
            
        
```

```python
round_output(y_pred)
```

```python
y_hat
```

```python
from sklearn import metrics
import seaborn as sns
```

```python
metrics.accuracy_score(Y_test,y_hat)
```

```python
cm_LR = metrics.confusion_matrix(Y_test,y_hat)
```

```python
sns.heatmap(cm_LR,annot=True,fmt='g',
           xticklabels=['Yes','No'],
           yticklabels=['Yes','No'])
plt.ylabel('Prediction',fontsize=11)
plt.xlabel('Actual',fontsize=11)
plt.title('Confusion Matrix', fontsize = 15)
plt.show()
```
